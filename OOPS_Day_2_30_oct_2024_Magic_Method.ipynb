```python
class fraction:
    
    def __init__(self , numerator , denomenator):
        
        self.numerator = numerator               # instance variable
        self.denomenator = denomenator           # instance variable 
        
        ## self is variable it can access instance variable and instance method
        ## to access num and deno , we use self .
        
    
    ## magic method : the method starts with __ and ends with __ i.e. magic or dunder method.
    
    def __str__(self):
        
        return "{} / {}".format(self.numerator , self.denomenator)
    
    
    # magic method
    
    # Here self is not by default . Here is self is parameter to store the value of (obj1)
    def __add__(self , other):
        
        ## self in the __add__ method refers to the first fraction (obj1), which holds 4/3
        ## other refers to the second fraction (obj2), which holds 2/3
        
        new_num = self.numerator * other.denomenator + other.numerator * self.denomenator
        new_den = self.denomenator * other.denomenator
        return "{} / {}".format(new_num , new_den)
    
    
    # magic method
    
    def __sub__(self , other):
        
        new_num = self.numerator * other.denomenator - other.numerator * self.denomenator
        new_den = self.denomenator * other.denomenator
        return "{} / {}".format(new_num , new_den)
    
    
    # magic method
    
    def __mul__(self , other):
        
        new_num = self.numerator * other.numerator
        new_den = self.denomenator * other.denomenator
        return "{} / {}".format(new_num , new_den)
    
    # magic method
    
    def __truediv__(self , other):
        
        new_num = self.numerator * other.denomenator
        new_den = self.denomenator * other.numerator
        return "{} / {}".format(new_num , new_den)
    
    
    # normal method
    
    def convert_to_decimal(self):
        
        return self.numerator / self.denomenator
        
    
    
### After call class , control goes at start and starting execute constr , pass arguments of class
### at time of calling pass to num and denom (at time def of constr) and goes that values to
### num and denom and we acess that values(variables) by self variable 
```

```python
class Rectangle:
    
    def __init__(self , length , width):
        
        self.length = length
        self.width = width
        
    def perimeter(self):
        
        perimeter_rectangle = 2 * (self.length + self.width)
        print('Perimeter of rectangle :', perimeter_rectangle)
        
    def area(self):
        
        area_rectangle = self.length * self.width
        print('Area of rectangle :' , area_rectangle)
        
    def display(self):
        
        print('Length of rectangle :' , self.length)
        print('Width of rectangle :' , self.width)
        self.perimeter()
        self.area()
        
        
```

```python
s = Rectangle(10 , 20)
```

```python
class rectangle:
    
    
    def __init__(self , l , w):
        
        self.__l = l            # __l is  make it as private
        self.__w = w            # __w is make it as public
                                # self is can acces the instance variables like l and w
            
            
    def __perimeter(self):        # __perimeter is make it as private method
        
        return 2*(self.__l + self.__w)
    
    
    def __area(self):            # __area is make it as private method
        
        return self.__l * self.__w
    
    
    def display(self):           # display is normal method
        
        print('Length of rectangle is :' , self.__l)
        print('Width of rectangle is :' , self.__w)
        print('Perimeter of rectangle is :', self.__perimeter())
        print('Area of rectnagle is :',self.__area())      
```

```python
class BankAccount():
    
    
    def __init__(self , accountNumber , name , balance):
        
        self.accountNumber = accountNumber
        self.name = name
        self.balance = balance
        
        
        
   
    def Deposit(self , deposit_amount):
        
        
        self.balance = self.balance + deposit_amount
        
        print('Deposite amount succesfully deposited....')
        print(f'{deposit_amount} amount succesfully deposited and  now balance is :{self.balance}'.format(deposit_amount , self.balance))
        
    #deposit_amount = int(input('Enter the deposit amount'))
    
    
    
    def withdrawl(self , withdraw_amount):
        
        if withdraw_amount > self.balance :
            
            print('Insufficient Balance....')
            
        else:
            
            self.balance = self.balance - withdraw_amount
            
            print(f'{withdraw_amount} amount withdraw successfully and now balance is :{self.balance}'.format(self.balance))
            
            
    #withdraw_amount = int(input('Enter the withdraw amount :' , ))
            
            
            
    def bankFees(self):
        
        
        bankfees = (self.balance * 0.05)              # 5 % fees of total balance
        
        return bankfees
    
    def display(self):
        
        print('BankAccount number is :' , self.accountNumber)
        print('Name of customer is :' , self.name)
        print('Balance is :' , self.balance)
        print('Bank fees are :' , self.bankFees())
        
```
