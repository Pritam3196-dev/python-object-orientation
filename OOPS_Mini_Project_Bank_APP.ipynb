### Build simple Application for Bank.

```python
class Bank:
    
    bankname = 'Bank of America'    # static variable
    branch_address = 'Charlotte, North Carolina, United States'  # static variable
    
    def __init__(self , name , greencard , address):
        
        self.name = name
        self.greencard = greencard
        self.address = address
        print( self.name , 'Your Account created successfully in',Bank.bankname,'in',Bank.branch_address , 'branch')
        self.balance = 0    # initially balance is zero at time creation of bank account
        
        
    
    def Deposite(self , amount):
        
        self.balance = self.balance + amount
        print('$', amount , 'Amount Deposited Successfully')
        
    
    def Withdraw(self , amount):
        
        if amount > self.balance :
            
            print('Insufficient Balance')
            
        else:
            
            self.balance = self.balance - amount
            
            print('$' , amount,'Amount get withdrawn from your account and now balance is in your account $:',self.balance)
            
        
    def Ministatement(self):
        
        print('Balance in your account is $:' , self.balance)
        
        
        



name = input('Enter your name for account creation :')
greencard = input('Enter the greencard number for account creation :')
address = input('Enter the address for account creation :')

#print('Name of Bank is :' , Bank.bankname , 'and' , 'Bank address is:' , Bank.branch_address)

b = Bank(name , greencard , address) #this brackets for calling constructor
# b.Deposite(100)
# b.Ministatement()
# b.Withdraw(50)
# b.Ministatement()

print('Welcome in ' , b.bankname)
print('press 1 for Deposite \npress 2 for Withdraw \npress 3 for Ministatement \npress 4 for stop')

while True : # infinite loop it get stops when it get false or explicitely break
    
    option = int(input('Enter Your Preferance Here:'))
    
    if option == 1:
        
        amount = int(input('Enter the amount for Deposite :'))
        b.Deposite(amount)
        
    elif option == 2:
        
        amount = int(input('Enter the amount for withdraw money :'))
        b.Withdraw(amount)
        
    elif option == 3:
        
        b.Ministatement()
        
    elif option == 4:
        print('Thank You For Associated With Us and Please Visit Again.....')
        break  # It get stops iteration and get control get out of loop(while loop i.e.stop)
        # we intentially terminate while loop means iteration or asking perferance
        
    else:
        print('Please Enter The Valid Input For Acessing The Services')
        # it again ask for input preferance
        # after execute else there is no code belowe else so again control goes to whilw loop
        # because still we inside the loop and while loop is infinite loop it will runs 
        # infinity times untill get false. 
```
