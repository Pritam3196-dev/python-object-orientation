#### Abstraction :
##### Abstraction is the data hiding process where we hide unnecessary details from user and it showing essential features to the user . or it hide the complexity of code and showing essential features to the user

```python
class car :
    
    def __init__(self):
        
        self.clutch = False      # False means we are not press clutch
        self.acc = False         # False means we are not press accelrature
        self.breakk = False      # False means we are not press breakk
        
    
    def car_start(self):
        
        self.clutch = True         # True means we are press clutch for starting car
        self.acc = True            # True means we are press accelreature for starting car
        
        print('car start..........')
```

```python
c = car()
```

```python
c.car_start()
```

##### Here as a developer we know what we did initially we are clutch , acc and breakk as False and then we make it as True . but the user is not aware about it he / she knows onlycar start means we are hiding unneccessary details or hising complexity of code and showing essential feature to the user like here user concern with car start only

```python
class Account:
    
    def __init__(self  , balance , account_no):
        
        self.account_no = account_no
        self.balance = balance
        
        
    def debit(self , debit_amount):
        
        #self.debit_amount = debit_amount
        
        if self.balance < debit_amount : 
            
            print('Insufficient Balance for withdraw . please insert valid amount which is less than your balance')
            
        else:
            
            self.balance = self.balance - debit_amount
        
            print(f'{debit_amount} amount successfully debited')
            
            self.get_balance()
        
    def get_credit(self , credit_amount):
        
        self.balance = self.balance + credit_amount 
        
        print(f' {credit_amount} amount successfully credited')
        self.get_balance()
        
    def get_balance(self):
        
        print(f'Balance is :  {self.balance}')
        
```

```python
a = Account (10000 , 1234)
```

```python
a.debit(2000)
```

```python
a.get_credit(1000)
```
