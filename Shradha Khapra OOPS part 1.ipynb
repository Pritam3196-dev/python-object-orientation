### Static Method

```python
class students:
    
    
    def __init__(self , name , marks):
        
        self.name = name
        self.marks = marks
        
    
    @staticmethod
    def hello(): 
        
        print('Hello Students....')    # static methods are those methods where self , 
                                       # instance and class variables
                                       # are not contains . we are defined without self 
                                       # but before method we need to defined @staticmethod decorator 
        
        
        
    def get_avg(self):
        
        summ = 0
        
        for mark in self.marks:
            
            summ = summ + mark
            
            
        avg = summ / len(self.marks)
        
        #return avg
        print(f'Hi  {self.name} , Your avg of marks is : {avg}')
```

```python
s1 = students('karan' , [97 , 98 , 99])
```

Here's a clear breakdown of your understanding:

- **`self` is used for instance variables**: When you define instance methods, you use `self` because it allows access to instance variables (attributes) that are specific to an object (instance) of the class. Since a static method **does not** rely on instance-specific data, there is no need for `self`.

- **`cls` is used for class variables**: In class methods, you use `cls` to access and modify class variables (attributes that are shared across all instances of the class). A static method **does not** operate on class data, and it is independent of any class or instance state. Hence, there's no need for `cls`.

So, when defining a static method:
- **No `self`**: Since you're not accessing instance-specific data.
- **No `cls`**: Since you're not accessing class-specific data.

That’s why static methods are independent of both the **instance** and the **class**, and you use the `@staticmethod` decorator to mark it.
